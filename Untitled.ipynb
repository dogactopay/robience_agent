```python
import json
```

```python
with open('test.json') as f:
     response = json.load(f)
```

```python
edges = response[0]['scenario']['content']['edges']
nodes = response[0]['scenario']['content']['nodes']

initial = list(set([z['id'] for z in response[0]['scenario']['content']['nodes']])-set([z['target'] for z in response[0]['scenario']['content']['edges']]))[0]

vp = [initial]
zt = [z for z in nodes if z['id']==initial]


for t in vp:
    for k in edges:
        if k['source']==t:
            vp.append(k['target'])
            zt.append([z for z in nodes if z['id']==k['target']][0])

print(vp)
```
